```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('obesity_dataset.csv')  # https://mitu.co.in/dataset
```

```python
df
```

##### Separate the input data

```python
x = df[['Height','Weight']]
```

```python
x
```

```python
plt.title('Unclustered Data')
plt.grid()
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x['Height'], x['Weight'])
```

##### Find the SSE / Inertia

```python
# import the class
from sklearn.cluster import KMeans
```

```python
# create the object
km = KMeans(n_clusters=10, random_state=0)
```

```python
# get the lables
labels = km.fit_predict(x)
```

```python
labels
```

```python
km.inertia_
```

```python
sse = []
for i in range(1,15):
    km = KMeans(random_state=0, n_clusters=i)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
plt.figure(figsize=(16,9))
plt.title('Elbow Method')
plt.grid()
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.xticks(range(1,15))
plt.plot(range(1,15), sse, marker='o')
```

##### Build the model

```python
km = KMeans(random_state=0, n_clusters=3)
```

```python
# predict the labels
labels = km.fit_predict(x)
```

```python
labels
```

```python
km.labels_
```

```python
cents = km.cluster_centers_
```

```python
cents
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Unclustered Data')
plt.grid()
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x['Height'], x['Weight'])

plt.subplot(1,2,2)
plt.title('Clustered Data')
plt.grid()
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x['Height'], x['Weight'], c = labels)
plt.scatter(cents[:,0], cents[:,1], marker = '*', s = 350, color='red')
```
